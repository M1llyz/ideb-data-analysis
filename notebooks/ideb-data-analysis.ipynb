```python
# Apenas identificar o nome correto das colunas disponíveis
print(df.columns.tolist())
```

```python
# Seleciona as colunas de Aprovação
colunas_taxas = [col for col in df.columns if 'VL_APROVACAO' in col]

# Filtra apenas as colunas de taxa de aprovação dos anos desejados
anos_desejados = ['2017', '2019', '2021']
colunas_interesse = [col for col in colunas_taxas if any(ano in col for ano in anos_desejados)]

# Selecionando colunas que identificam os municipios e a rede específica
colunas_base = ['SG_UF', 'NO_MUNICIPIO', 'REDE']

# Junta as colunas de interesse e as de base (munícipios e rede)
df_filtrado = df[colunas_base + colunas_interesse]

df_filtrado
```

```python
# Agrupando por ano e tipo de rede (Estadual, Municipal, Pública)
df_melt = df_filtrado.melt(
    id_vars=['NO_MUNICIPIO', 'REDE', 'SG_UF'],
    value_vars=colunas_interesse,
    var_name='Ano_Coluna',
    value_name='Taxa_Aprovacao'
)

# Usa regex (\d{4}) para extrair o ano (4 dígitos) da string |  VL_APROVACAO_2017_SI_4 -> 2017
df_melt['Ano'] = df_melt['Ano_Coluna'].str.extract(r'(\d{4})').astype(int)

# Convertendo a taxa de aprovação para numérico, se houver erro vira Nan(nulo)
df_melt['Taxa_Aprovacao'] = pd.to_numeric(df_melt['Taxa_Aprovacao'], errors='coerce')

# Remove os possíveis Nan(nulos) que apareceram
df_melt = df_melt.dropna(subset=['Taxa_Aprovacao'])
```

```python
# Filtra para municípios do estado de São Paulo (para evitar São Vicente-RN e Praia Grande-SC)
df_melt = df_melt[df_melt['SG_UF'] == 'SP']

# Lista das cidades da Baixada Santista
cidades_baixada = [
    "Bertioga", "Cubatão", "Guarujá", "Itanhaém",
    "Mongaguá", "Peruíbe", "Praia Grande", "Santos", "São Vicente"
]

# Filtra apenas os municípios da Baixada Santista
df_baixada = df_melt[df_melt['NO_MUNICIPIO'].isin(cidades_baixada)]
```

```python
# Filtra os dados apenas para os anos de interesse
df_anos_interesse = df_baixada[df_baixada['Ano'].isin([2017, 2019, 2021])]

# Calculando a média da taxa de aprovação por município e ano
df_media_por_municipio_ano = df_anos_interesse.groupby(['NO_MUNICIPIO', 'Ano'])['Taxa_Aprovacao'].mean().reset_index()

df_media_por_municipio_ano
```

```python
import plotly.express as px

# Gera gráfico de linha
fig = px.line(
    df_media_por_municipio_ano,
    x='Ano',
    y='Taxa_Aprovacao',
    color='NO_MUNICIPIO',
    markers=True,
    title='Evolução da Taxa de Aprovação Por Município - Baixada Santista (2017, 2019, 2021)'
)

# Ajustes estéticos
fig.update_layout(
    template='simple_white',
    title_font_size=20,
    title_x=0.5,
    legend_title_text='Município',
    legend=dict(
        title_font_size=12,
        font_size=10,
        bordercolor="Black",
        borderwidth=0.5
    ),
    xaxis_title='Ano Letivo',
    yaxis_title='Taxa de Aprovação (%)',
    xaxis=dict(
        tickmode='array',
        tickvals=[2017, 2019, 2021]
    )
)
fig.update_traces(line=dict(width=2), marker=dict(size=6), opacity=0.8)

fig.show()
```

```python
# Calculando a média da taxa de aprovação para toda a Baixada Santista por ano e rede
df_media_baixada_rede = df_baixada.groupby(['Ano', 'REDE'])['Taxa_Aprovacao'].mean().reset_index()

df_media_baixada_rede
```

```python
import plotly.express as px

# Gera o gráfico de linha
fig2 = px.line(
    df_media_baixada_rede,
    x='Ano',
    y='Taxa_Aprovacao',
    color='REDE',
    markers=True,
    title='Evolução da Taxa de Aprovação Média por Rede - Baixada Santista (2017, 2019, 2021)'
)

# Ajustes estéticos
fig2.update_layout(
    template='simple_white',
    title_font_size=20,
    title_x=0.5,
    legend_title_text='Rede de Ensino',
    legend=dict(
        title_font_size=14,
        font_size=12,
        bordercolor="Black",
        borderwidth=0.5,
    ),
    xaxis_title='Ano Letivo',
    yaxis_title='Taxa de Aprovação (%)',
    xaxis=dict(
        tickmode='array',
        tickvals=[2017, 2019, 2021]
    )
)
fig2.update_traces(line=dict(width=2), marker=dict(size=6), opacity=0.8)

fig2.show()
```
